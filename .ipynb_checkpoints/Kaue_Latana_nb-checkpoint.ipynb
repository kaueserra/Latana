# Fixing missing values 

There are three variables to be fixed missing values ( dem.education_level, question.work_start_up, question.personal_area_of_work), however the personal_area_of_work besides having a high percentage of 
missing values also has too many categories, therefore the treatment for this feature will be creating a category name "missing",  the other two will be used MICE algorithm to impute the missing values.  

For more info about MICE, the paper below:

https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3074241/



```python
non_cat = ['meta.uuid','dem.age'] 

for i in ds.columns: 
    ds[i] = ds[i].astype('category')  
    
ds.index = ds['meta.uuid']  

ds = ds.drop('meta.uuid',axis=1)  

X = ds.drop('question.brand_awareness_client',axis=1)  


Y = ds['question.brand_awareness_client']
```

```python
# To avoid leakeage the imputation will be made without the brand_awareness_client feature

kds = mf.KernelDataSet(
  X,
  save_all_iterations=True,
  random_state=42
)

# Run the MICE algorithm for 3 iterations
kds.mice(10)

# Return the completed kernel data
ds_noms = kds.complete_data()
```

## Getting the best features 


```python
# Creating dumies for features with more than two categories and transforming str into numeric

ds_noms['question.brand_awareness_client'] = ds['question.brand_awareness_client']

unique_counts = pd.DataFrame.from_records([(col, ds_noms[col].nunique()) for col in ds_noms.columns],
                          columns=['Column_Name', 'Num_Unique']).sort_values(by=['Num_Unique'])

label_enc_feat = unique_counts[unique_counts.Num_Unique==2]['Column_Name'].values 

oneh_enc_feat = unique_counts[(unique_counts.Num_Unique>2) & (unique_counts.Num_Unique<40)]['Column_Name'].values  

labelencoder = LabelEncoder()  

onehotencoder = OneHotEncoder()   

ds_noms1 =ds_noms.copy() 
```

```python
# Joining the transformed datasets, transforming everything in integer and fixing feature names that might give some error

ds_noms2=  pd.get_dummies(ds_noms1[oneh_enc_feat],drop_first=True) 

ds_noms1[label_enc_feat] = ds_noms1[label_enc_feat].apply(LabelEncoder().fit_transform)

final_ds = pd.merge(ds_noms1,ds_noms2,how='left',left_index=True,right_index=True) 

final_ds = final_ds.drop(oneh_enc_feat,axis=1)  

final_ds = final_ds.astype('int32')

final_ds = final_ds.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '_', x)) 
```

```python
# As we got a lot of dummies better check and drop duplicated columns 

print(final_ds.shape) 

final_ds = final_ds.T.drop_duplicates().T 

print(final_ds.shape) 

# 10 repeated columns 
```

```python
#Split the data into train and test   

xtrain = final_ds.drop('question_brand_awareness_client',axis=1)

labels = final_ds['question_brand_awareness_client']

x_train, x_test, y_train, y_test = train_test_split(xtrain,labels,test_size = 0.25,train_size =0.75,random_state=42)
```

```python
#sm = SMOTE(random_state=42) 

#x_train_res, y_train_res = sm.fit_resample(x_train, y_train) 
```

```python
# I will run a Random Forest to use as baseline and select features, with feature importance 

rf = RandomForestClassifier(n_estimators=1000, random_state=42,class_weight='balanced',min_samples_leaf=10)
rf.fit(x_train,y_train)
print('-*-'*20)



predictions_train_RF = rf.predict(x_train)
# Use accuracy_score function to get the accuracy
print("RF train F1 Score -> ",f1_score(predictions_train_RF, y_train,average='macro'))


print('-*-'*20)


predictions_test_RF = rf.predict(x_test)
# Use accuracy_score function to get the accuracy
print("RF test F1 Score -> ",round(f1_score(predictions_test_RF, y_test,average='macro')*100,2))  

print(classification_report(predictions_test_RF, y_test))

# We could notice a overfiting from the RF, probably because there was no prunning or tunning,  
# Traning was hitting 100% of f1 score, then I introduced minimum leaf size to 10
```

```python
#plotting only features with > 0.1% of importance
feature_imp_total = pd.DataFrame(sorted(zip(rf.feature_importances_,xtrain.columns)), columns=['Value','Feature'])

feature_imp =feature_imp_total[feature_imp_total.Value > 0.01]
    
plt.figure(figsize=(20, 10))
sns.barplot(x="Value", y="Feature", data= feature_imp.sort_values(by="Value", ascending=False))
plt.title('Feature Importance')
plt.tight_layout()
plt.show()
```

```python
top_feat  = list(feature_imp_total[feature_imp_total.Value > 0.01]['Feature']) 

lda_feat = list(feature_imp_total[(feature_imp_total.Value < 0.01) & (feature_imp_total.Value > 0.001) ]['Feature']) 
```

```python
# Creating a feature with all the features that explain less than 1% and more than 0.1%  

lda = LinearDiscriminantAnalysis(n_components=1)

x_train['lda_disc'] = lda.fit(final_ds[lda_feat], labels).transform(x_train[lda_feat])   

x_test['lda_disc'] = lda.fit(final_ds[lda_feat], labels).transform(x_test[lda_feat]) 

top_feat.append('lda_disc')
```

```python
rf = RandomForestClassifier(n_estimators=10000, random_state=42,class_weight='balanced', min_samples_leaf=10)
rf.fit(x_train[top_feat],y_train)

print('-*-'*20)



predictions_train_RF = rf.predict(x_train[top_feat])
# Use accuracy_score function to get the accuracy
print("RF train F1 Score -> ",f1_score(predictions_train_RF, y_train,average='macro'))


print('-*-'*20)


predictions_test_RF = rf.predict(x_test[top_feat])
# Use accuracy_score function to get the accuracy
print("RF test F1 Score -> ",round(f1_score(predictions_test_RF, y_test,average='macro')*100,2))  

print(classification_report(predictions_test_RF, y_test))


# We could notice a overfiting from the RF, probably because there was no prunning or tunning
```

```python
# Checking if there was any change on the feature importance 
feature_imp_total1 = pd.DataFrame(sorted(zip(rf.feature_importances_,top_feat)), columns=['Value','Feature'])

    
plt.figure(figsize=(20, 10))
sns.barplot(x="Value", y="Feature", data= feature_imp_total1.sort_values(by="Value", ascending=False))
plt.title('Feature Importance')
plt.tight_layout()
plt.show()
```

```python
%%time

def bayes_parameter_opt_lgb(X, y, init_round=15, opt_round=25, n_folds=3, random_seed=6,n_estimators=10000, output_process=False):
    # prepare data
    train_data = lgb.Dataset(data=X, label=y, free_raw_data=False)
    # parameters
    def lgb_eval(learning_rate,num_leaves, feature_fraction, max_depth, max_bin, min_data_in_leaf,min_sum_hessian_in_leaf,subsample):
        params = {'application':'binary', 'metric':'average_precision'}
        params['learning_rate'] = max(min(learning_rate, 1), 0)
        params["num_leaves"] = int(round(num_leaves))
        params['feature_fraction'] = max(min(feature_fraction, 1), 0)
        params['max_depth'] = int(round(max_depth))
        params['max_bin'] = int(round(max_depth))
        params['min_data_in_leaf'] = int(round(min_data_in_leaf))
        params['min_sum_hessian_in_leaf'] = min_sum_hessian_in_leaf
        params['subsample'] = max(min(subsample, 1), 0)
        
        cv_result = lgb.cv(params, train_data, nfold=n_folds, seed=random_seed,verbose_eval =400, stratified=True, metrics=['average_precision'])
        return max(cv_result['average_precision-mean'])
     
    lgbBO = BayesianOptimization(lgb_eval, {'learning_rate': (0.01, 1.0),
                                            'num_leaves': (24, 80),
                                            'feature_fraction': (0.1, 0.9),
                                            'max_depth': (5, 30),
                                            'max_bin':(20,90),
                                            'min_data_in_leaf': (20, 80),
                                            'min_sum_hessian_in_leaf':(0,100),
                                           'subsample': (0.01, 1.0)}, random_state=200)

    
    #n_iter: How many steps of bayesian optimization you want to perform. The more steps the more likely to find a good maximum you are.
    #init_points: How many steps of random exploration you want to perform. Random exploration can help by diversifying the exploration space.
    
    lgbBO.maximize(init_points=init_round, n_iter=opt_round)
    
    model_auc=[]
    for model in range(len( lgbBO.res)):
        model_auc.append(lgbBO.res[model]['target'])
    
    # return best parameters
    return lgbBO.res[pd.Series(model_auc).idxmax()]['target'],lgbBO.res[pd.Series(model_auc).idxmax()]['params']

opt_params = bayes_parameter_opt_lgb(x_train[top_feat],y_train, init_round=5, opt_round=10, n_folds=3, random_seed=6,n_estimators=10000)
```

```python
opt_params[1]['num_leaves'] = int(round(opt_params[1]['num_leaves'])) 
opt_params[1]['max_depth'] = int(round(opt_params[1]['max_depth']))  
opt_params[1]['max_bin'] = int(round(opt_params[1]['max_bin'])) 
opt_params[1]['min_data_in_leaf'] = int(round(opt_params[1]['min_data_in_leaf'])) 

opt_params[1]
```

```python
lgbm = lgb.LGBMClassifier( num_iterations =1000,class_weight='balanced',
    **opt_params[1]
) 

lgbm.fit(x_train[top_feat],y_train) 


predictions_train_lgbm = lgbm.predict(x_train[top_feat])
# Use accuracy_score function to get the accuracy
print("RF test F1 Score -> ",round(f1_score(predictions_train_lgbm, y_train,average='macro')*100,2) )

print('-*-'*20)


predictions_test_lgbm = lgbm.predict(x_test[top_feat])
# Use accuracy_score function to get the accuracy
print("RF val F1 Score -> ",round(f1_score(predictions_test_lgbm, y_test,average='macro')*100,2) )

print('-*-'*20) 

print(classification_report(predictions_test_lgbm, y_test))
```
